```python
## sci
for dataset in combine:
    
##1번 대출 총 횟수
    dataset['LNIF_CNT_SUM'] = dataset['BNK_LNIF_CNT'] + dataset['CPT_LNIF_CNT'] + dataset['SPART_LNIF_CNT'] + dataset['ECT_LNIF_CNT']

#2,3,4,5번  #대출 부문당 비율
    dataset['LNIF_CNT_RATE_1'] = dataset['BNK_LNIF_CNT'] / dataset['LNIF_CNT_SUM'] 
    dataset['LNIF_CNT_RATE_2'] = dataset['CPT_LNIF_CNT'] / dataset['LNIF_CNT_SUM'] 
    dataset['LNIF_CNT_RATE_3'] = dataset['SPART_LNIF_CNT'] / dataset['LNIF_CNT_SUM'] 
    dataset['LNIF_CNT_RATE_4'] = dataset['ECT_LNIF_CNT'] / dataset['LNIF_CNT_SUM'] 

# 6번 평균대출금액
    dataset['AVG_LNIF_AMT'] = dataset['TOT_LNIF_AMT'] / dataset['LNIF_CNT_SUM']

# 7번 1회평균 은행대출금액
    dataset['AVG_BNK_AMT'] = dataset['BNK_LNIF_AMT'] / dataset['BNK_LNIF_CNT']

# 8번 1회평균 카/할/캐 대출금액 
    dataset['AVG_CPT_AMT'] = dataset['CPT_LNIF_AMT'] / dataset['CPT_LNIF_CNT']

# 9번 모든 금융권 신용대출건수 유지기간 총합
    dataset['TOT_CLIF_MDIF'] = dataset['CRDT_OCCR_MDIF'] + dataset['SPTCT_OCCR_MDIF']

# 10번 신용카드개수당 총 유지기간
    dataset['TOT_CTCD_MDIF'] = dataset['CRDT_CARD_CNT'] * dataset['CTCD_OCCR_MDIF']

# 11번 평균보증금액 
    dataset['AVG_GUIF_AMT'] = dataset['CB_GUIF_AMT'] / dataset['CB_GUIF_CNT']

# 12번 은행대출건수(제곱)- 카/할/캐 대출건수 - 2산업분류 대출건수 - 기타 대출건수
    dataset['loan_count1'] = (dataset['BNK_LNIF_CNT']+1)**2 - dataset['CPT_LNIF_CNT'] - dataset['SPART_LNIF_CNT'] - dataset['ECT_LNIF_CNT']

# 13번 은행대출건수 - 카/할/캐 대출건수 - 2산업분류 대출건수 - 기타 대출건수
    dataset['loan_count2'] = dataset['BNK_LNIF_CNT'] - dataset['CPT_LNIF_CNT'] - dataset['SPART_LNIF_CNT'] - dataset['ECT_LNIF_CNT']

# 14번 (카/할/캐 대출건수 + 2산업분류 대출건수 + 기타 대출건수) / 은행대출건수
    dataset['loan_count3'] = (dataset['BNK_LNIF_CNT'] + dataset['SPART_LNIF_CNT'] + dataset['ECT_LNIF_CNT']) / dataset['BNK_LNIF_CNT']

# 15번 보증 금액 / 대출금액 
    dataset['GUIF_LNIF_RATE'] = dataset['CB_GUIF_AMT'] / dataset['TOT_LNIF_AMT']

# 16번 전체에서 은행 외 대출금액 비율
    dataset['LNIF_RATE_EXB'] = (dataset['TOT_CLIF_AMT'] + dataset['CPT_LNIF_AMT']) / dataset['TOT_LNIF_AMT']

# 17,18,19번총 대출금액 중 총 신용대출금액 /은행대출금액 비율 / 카/할/캐대출 금액 비율
    dataset['RATIO_CLIF_AMT'] =  dataset['TOT_CLIF_AMT'] / dataset['TOT_LNIF_AMT']
    dataset['RATIO_BNK2TOT_AMT'] =  dataset['BNK_LNIF_AMT'] / dataset['TOT_LNIF_AMT']
    dataset['RATIO_CPT2TOT_AMT'] =  dataset['CPT_LNIF_AMT'] / dataset['TOT_LNIF_AMT']

# 20번 은행 분의 나머지
    dataset['LNIF_RATE_4'] = (dataset['TOT_CLIF_AMT']+dataset['CPT_LNIF_AMT']) / (dataset['BNK_LNIF_AMT'])

# 21번 신용대출 분의 카드사/할부사/캐피탈x
    dataset['LNIF_RATE_5'] = (dataset['CPT_LNIF_AMT']) / (dataset['TOT_CLIF_AMT'])

# 22,23,24,25,26,27번 은행에서의 대출건과 나머지들 비교 변수
    dataset['LNIF_CNT_COMP_1'] = dataset['BNK_LNIF_CNT'] - dataset['CPT_LNIF_CNT']
    dataset['LNIF_CNT_COMP_2'] = dataset['BNK_LNIF_CNT'] - dataset['SPART_LNIF_CNT']
    dataset['LNIF_CNT_COMP_3'] = dataset['BNK_LNIF_CNT'] - dataset['ECT_LNIF_CNT']
    dataset['LNIF_CNT_COMP_4'] = dataset['BNK_LNIF_CNT']*2 - dataset['CPT_LNIF_CNT']
    dataset['LNIF_CNT_COMP_5'] = dataset['BNK_LNIF_CNT']*2- dataset['SPART_LNIF_CNT']
    dataset['LNIF_CNT_COMP_6'] = dataset['BNK_LNIF_CNT']*2 - dataset['ECT_LNIF_CNT']

# 28번 신용대출 분의 은행 비율
    dataset['BNK_CLIF_RATE'] = dataset['BNK_LNIF_AMT'] / (dataset['TOT_CLIF_AMT'])
```
